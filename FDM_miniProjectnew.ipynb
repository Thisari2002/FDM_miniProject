```python

# Identifying outliers in 'LN_IC50' using interquartile range (IQR)
Q1 = data['LN_IC50'].quantile(0.25)
Q3 = data['LN_IC50'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Removing outliers
data = data[~((data['LN_IC50'] < (Q1 - 1.5 * IQR)) |
                                  (data['LN_IC50'] > (Q3 + 1.5 * IQR)))]
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder


# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# List of categorical columns to convert to numeric
categorical_columns = [
    "CELL_LINE_NAME", "TCGA_DESC", "DRUG_NAME",
    "GDSC Tissue descriptor 1", "GDSC Tissue descriptor 2",
    "Cancer Type (matching TCGA label)", "Microsatellite instability Status (MSI)",
    "Screen Medium", "Growth Properties", "CNA",
    "Gene Expression", "Methylation", "TARGET", "TARGET_PATHWAY"
]

# Apply label encoding for all categorical columns using .loc[]
for col in categorical_columns:
    if col in data.columns:  # Check if the column exists in the DataFrame
        data.loc[:, col] = label_encoder.fit_transform(data[col].astype(str))


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Set title and labels
plt.title('Correlation Matrix', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Step 1: Select upper triangle of the correlation matrix
upper = correlation_matrix.abs().where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

# Step 2: Find features with correlation greater than the threshold
threshold = 0.8
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

print("Features to drop due to high correlation:", to_drop)

```

```python
from sklearn.feature_selection import SelectKBest, f_regression
import pandas as pd

# Assuming 'LN_IC50' is your target variable
y = data['LN_IC50']  # Target variable

# Assuming all other columns except 'LN_IC50' are features
X = data.drop(columns=['LN_IC50'])

# Select top 10 features based on univariate regression tests
selector = SelectKBest(score_func=f_regression, k=5)  # Selecting 4 features first
selector.fit(X, y)

# Get the selected features
selected_features = X.columns[selector.get_support()]
print("Selected Features:", selected_features)

# Manually add the 'LN_IC50' column to selected features for analysis
selected_features_with_ic50 = list(selected_features) + ['LN_IC50']

# Subset the data to only include the selected features + IC50
X_selected_with_ic50 = data[selected_features_with_ic50]

```

```python
# Rescaling to [0, 1] range
cleaned_data['LN_IC50_rescaled'] = (cleaned_data['LN_IC50'] - cleaned_data['LN_IC50'].min()) / (cleaned_data['LN_IC50'].max() - cleaned_data['LN_IC50'].min())

# Applying Logit Transformation
cleaned_data['LN_IC50_logit_transformed'] = np.log(cleaned_data['LN_IC50_rescaled'] / (1 - cleaned_data['LN_IC50_rescaled']))

# Plot Histogram
plt.figure(figsize=(12, 10))
sns.histplot(cleaned_data['LN_IC50_logit_transformed'], bins=30, kde=True)
plt.title('Logit Transformation of Rescaled LN_IC50')
plt.xlabel('Logit Transformed LN_IC50')
plt.ylabel('Frequency')
plt.show()
```

# **Model Training**

```python
#Data Splitting
from sklearn.model_selection import train_test_split
```

```python
X = cleaned_data.drop('LN_IC50', axis=1)  # Features
y = cleaned_data['LN_IC50']  # Target variable
```

```python
# Splitting the data (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Verify split sizes
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

**Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, mean_squared_error
```

```python
# Initialize the Linear Regression model
linear_model = LinearRegression()

# Fit the model on training data
linear_model.fit(X_train[['CELL_LINE_NAME', 'DRUG_NAME', 'AUC', 'Z_SCORE', 'TARGET_PATHWAY']], y_train)

# Make predictions on the test data
y_pred_reg = linear_model.predict(X_test[['CELL_LINE_NAME', 'DRUG_NAME', 'AUC', 'Z_SCORE', 'TARGET_PATHWAY']])
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Evaluate the model
# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred_reg)

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred_reg)

# Calculate R² score (Coefficient of Determination)
r2 = r2_score(y_test, y_pred_reg)

# Print all regression metrics
print(f'Linear Regression - MSE: {mse}')
print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
print(f'R²: {r2}')
```

```python
import matplotlib.pyplot as plt

# Regression metrics
metrics = {
    'MSE': mse,
    'RMSE': rmse,
    'MAE': mae,
    'R²': r2
}

# Create bar graph
plt.figure(figsize=(10, 6))
plt.bar(metrics.keys(), metrics.values(), color=['blue', 'orange', 'green', 'red'])

# Add titles and labels
plt.title('Regression Metrics for Linear Regression Model', fontsize=16)
plt.xlabel('Metrics', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Show grid for better readability
plt.grid(axis='y')

# Show the plot
plt.show()
```

# **HperParameter Tunning - Random Forest Model**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define parameter grid
param_grid = {
    'n_estimators': [100],
    'max_depth': [None],
    'min_samples_split': [2],
    'min_samples_leaf': [1]
}

# Initialize Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV with 5-fold cross-validation
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)

# Fit the model on training data
grid_search.fit(X_train[['CELL_LINE_NAME', 'DRUG_NAME', 'AUC', 'Z_SCORE', 'TARGET_PATHWAY']], y_train)

# Get the best estimator
best_rf_model = grid_search.best_estimator_

# Make predictions on the test data
y_pred_rf = best_rf_model.predict(X_test[['CELL_LINE_NAME', 'DRUG_NAME', 'AUC', 'Z_SCORE', 'TARGET_PATHWAY']])

# Evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
rmse_rf = np.sqrt(mse_rf)
mae_rf = mean_absolute_error(y_test, y_pred_rf)

# Print evaluation metrics
print(f'Random Forest with Hyperparameter Tuning - MSE: {mse_rf}, RMSE: {rmse_rf}, MAE: {mae_rf}, R²: {r2_rf}')
```

```python
# Predict on the training set using the best estimator
y_train_pred_rf = best_rf_model.predict(X_train[['CELL_LINE_NAME', 'DRUG_NAME', 'AUC', 'Z_SCORE', 'TARGET_PATHWAY']])

# Calculate MSE for the training set
mse_train_rf = mean_squared_error(y_train, y_train_pred_rf)

# Print the Training MSE
print(f'Training MSE: {mse_train_rf}')
```

```python
# Perform cross-validation on the best_rf_model
cv_scores = cross_val_score(
    best_rf_model, 
    X_train[['CELL_LINE_NAME', 'DRUG_NAME', 'AUC', 'Z_SCORE', 'TARGET_PATHWAY']], 
    y_train, 
    cv=5,  # 5-fold cross-validation
    scoring='neg_mean_squared_error'  # Use negative MSE to follow scikit-learn's convention
)

# Convert negative MSE scores to positive for interpretation
mse_scores = -cv_scores

# Calculate mean and standard deviation of MSE
mean_mse = np.mean(mse_scores)
std_mse = np.std(mse_scores)

# Print cross-validated MSE
print(f'Cross-Validated MSE: {mean_mse:.4f} ± {std_mse:.4f}')

# Feature importances
feature_importances = best_rf_model.feature_importances_
features = ['CELL_LINE_NAME', 'DRUG_NAME', 'AUC', 'Z_SCORE', 'TARGET_PATHWAY']
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances}).sort_values(by='Importance', ascending=False)
```
